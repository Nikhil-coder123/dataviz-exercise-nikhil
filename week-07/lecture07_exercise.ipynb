```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('netflix_catalogue.csv')
print(f"Loaded: {len(df)} titles")
print(df['type'].value_counts())
print(df.head())
```

```python
print("Genres:", df['genre'].value_counts().head(8))
print("\nCountries:", df['country'].value_counts().head(8))
print("\nRatings:", df['rating'].value_counts())
```

Task 1 — **Heatmap**

```python

df['decade'] = (df['release_year'] // 10 * 10).astype(str) + 's'


top_ratings = df['rating'].value_counts().nlargest(8).index

filtered = df[df['rating'].isin(top_ratings)]


rating_decade = (
    filtered.groupby(['rating', 'decade'])
    .size()
    .reset_index(name='count')
)


pivot = rating_decade.pivot(
    index='rating',
    columns='decade',
    values='count'
).fillna(0)


fig = px.imshow(
    pivot,
    text_auto=True,
    color_continuous_scale='YlOrRd',
    aspect='auto',
    title='Netflix Content by Rating and Release Decade'
)


fig.update_layout(
    width=1000,
    height=600,
    xaxis_title='Release Decade',
    yaxis_title='Content Rating',
    title_x=0.5
)

fig.show()
```

Task 2 — Waterfall **Chart** bold text

```python
# Import libraries
import pandas as pd
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv('netflix_catalogue.csv')

# Filter only Movies
movies = df[df['type'] == 'Movie']

# Group by added year
adds = (
    movies.groupby('added_year')
    .size()
    .reset_index(name='new_titles')
)

# Keep years between 2015 and 2022
adds = adds[
    (adds['added_year'] >= 2015) &
    (adds['added_year'] <= 2022)
]

# Calculate total movies added
cumulative = adds['new_titles'].sum()

# Create Waterfall Chart
fig = go.Figure(go.Waterfall(

    x=[str(year) for year in adds['added_year']] + ['Total'],

    y=adds['new_titles'].tolist() + [cumulative],

    measure=['relative'] * len(adds) + ['total'],

    text=adds['new_titles'].tolist() + [cumulative],

    textposition='outside',

    connector=dict(
        line=dict(
            color='#AAAAAA',
            dash='dot'
        )
    ),

    increasing=dict(
        marker_color='#70AD47'
    ),

    totals=dict(
        marker_color='#2E75B6'
    )
))

# Layout
fig.update_layout(

    title='Growth of Netflix Movie Library (2015–2022)',

    xaxis_title='Year',

    yaxis_title='Number of Movies Added',

    width=1000,

    height=600,

    title_x=0.5
)

# Show chart
fig.show()
```
